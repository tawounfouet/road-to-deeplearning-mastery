```python
import tensorflow as tf

mnist = tf.keras.datasets.mnist
(trainings_images, training_labels), (testing_images, testing_labels) = mnist.load_data()
print(f"Training Images Shape: {trainings_images.shape} &  Training Label Shape {training_labels.shape}")
print(f"Testing Images Shape: {testing_images.shape} &  Testing Label Shape {testing_labels.shape}")
```

```python
trainings_images[0]
```

```python
# reshaping
print(f"Shape before reshaping: {trainings_images.shape} ")
trainings_images = trainings_images.reshape(*trainings_images.shape, 1)
print(f"Shape after reshaping: {trainings_images.shape} ")

# normalization
print("First image beforme normalization")
print(trainings_images[0][1][1])
trainings_images = trainings_images / 255 # normalizing
print("First imgae beforme normalization")
print(trainings_images[0][1][1])
```

```python
# reshaping
print(f"Shape before reshaping: {testing_images.shape} ")
testing_images = testing_images.reshape(*testing_images.shape, 1)
print(f"Shape after reshaping: {testing_images.shape} ")

# normalization
print("First image beforme normalization")
print(testing_images[0][1][1])
testing_images = testing_images / 255 # normalizing
print("First image beforme normalization")
print(testing_images[0][1][1])
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), activation="relu", input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(10, activation = "softmax")
])
```

```python
# dimension = shape = (height, width)


# MNIST Images (28, 28, 1) :
# - Input_shape (28, 28)
# - filter/channel = 1
# - activation_shape = 28 * 28 = 784
# - Pararmeters = 0

# Conv2D (64, (3,3)
 # - A Filter shape (3, 3)  or (Filter = 3*3)
 # - number_of_filters (channels) = 64
 # - layer_neuron_shape = [input_dimensions - filters_dimension + (2 * padding) / strides] + 1
  #- layer_neuron_shape = [28 - 3 + (2 * 0) / 1 ] + 1 = 26
  # - activation_shape = height * width * number_of_filters
  # - activation_shape = 26 * 26 * 64
 # params =  [(filter_height * filter_width * num_filter_last_layer) + 1 ] * num_filters_current_layer
  # params = [(3 * 3 * 1) + 1] * 64 = 640

# MaxPooling2D (2, 2)
# - layer_neuron_shape = [(conv_2d_dimensions - pool_dimension) / Strides ] + 1
# - layer_neuron_shape = ((26 - 2) / 2) + 1  = 13  => (13, 13, 64 )
# - activation_shape = 13 * 13 * 64 = 10816
# - params = 0


# Flatten Layer ()
# layer_neuron shape (10816, 1)
# activation shape  = 10816
# params = 0

# Dense(128)
# - layer_neuron_shape (128, 1)
# - activation_shape = 128
# - params = [(current_layesr_neurons * previous_layer_neurons) + (1 * current_layers_neurons)]
# - params = (128 * 10816) + (1*128) = 1.384.576   # 1 represent the bias


# Dense(10)
# - shape (10, 1)
# - activation_shape = 10
# - params = (10 * 128) + (1 * 10) = 1290
```
