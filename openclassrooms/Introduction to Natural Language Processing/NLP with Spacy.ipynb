```python
doc = nlp(text)
sentence1 = list(doc.sents)[0]
sentence1
```

```python
import numpy as np
your_word = "Adventures"


ms = nlp.vocab.vectors.most_similar(
    np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=10)

words = [nlp.vocab.strings[w] for w in ms[0][0]]
distances = ms[2]
print(words)
```

```python
doc2 = nlp("I like salty fries and hamburgers.")
doc3 = nlp("Fast food tastes very good.")
```

```python
print(doc2, "<->", doc3, doc2.similarity(doc3))
```

```python
doc4 = nlp("The Empire State Buliding is in New York")
```
